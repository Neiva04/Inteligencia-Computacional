# 2.2b Regressão

## Previsão de valores contínuos



> ***O termo "regressão" vem da estatística e foi usado pela primeira vez por Francis Galton, primo de Charles Darwin, no século XIX. Galton estava estudando a relação entre as alturas de pais e filhos. Ele notou que filhos de pais muito altos ou muito baixos tendiam a "regredir" para a altura média da população. Daí o termo "regressão". Hoje, usamos o termo em um contexto mais amplo para falar sobre modelagem de relações entre variáveis.***
