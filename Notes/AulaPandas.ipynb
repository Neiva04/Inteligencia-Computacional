# O que é o Pandas?

Pandas é uma biblioteca open-source para Python utilizada para manipulação e análise de dados. Ela é construída em cima do NumPy e oferece estruturas de dados fáceis de usar e ferramentas de análise de dados. A biblioteca é especialmente útil para trabalhar com dados tabulares (como planilhas ou tabelas SQL) e dados de séries temporais.

As principais estruturas de dados no Pandas são  **Series** e o **DataFrame**. Um Series é um objeto unidimensional semelhante a um array que pode armazenar qualquer tipo de dado (inteiros, strings, floats, etc.) enquanto um DataFrame é uma estrutura de tabela bidimensional que é composta por vários Series.



image.png



O Pandas oferece uma ampla gama de funções para manipulação e análise de dados, incluindo limpeza de dados, agregação, filtragem, mesclagem e agrupamento. A biblioteca também fornece ferramentas para trabalhar com dados faltantes e para transformar dados em diferentes formatos.

Prós:

    Sintaxe fácil de usar e intuitiva para manipulação e análise de dados.
    Fornece ferramentas poderosas para trabalhar com dados tabulares e de séries temporais.
    Integra-se bem com outras bibliotecas Python como NumPy, Matplotlib e Scikit-Learn.
    Grande comunidade de suporte e desenvolvimento ativo, com atualizações e melhorias regulares.

Contras:

    Pode ser lento ao trabalhar com conjuntos de dados muito grandes.
    Pode ser intensivo em memória, especialmente ao trabalhar com vários DataFrames grandes.
    Algumas das funcionalidades mais avançadas podem ser difíceis de aprender e entender para iniciantes.

# Primeiros passos: obtendo um dataset

```python
import requests

target_url = "https://raw.githubusercontent.com/fivethirtyeight/data/master/nba-elo/nbaallelo.csv"
filename = "nba_elo.csv"

response = requests.get(target_url)
response.raise_for_status()    # Check that the request was successful
with open(filename, "wb") as f:
    f.write(response.content)
print(f"Arquivo {filename} está disponível")
```

```python
# Carregando lib
import pandas as pd

# Leia o arquivo em um dataframe do Pandas
df = pd.read_csv(filename)

# Ver a primeira linha do dataframe
#print(df.head())
# Ver a n linhas do dataframe
df.head(1)
# Ver a última linha do dataframe
# print(df.tail())
```

# Encontrando elementos específicos

## Label vs. Localização

A principal diferença entre eles:

    loc recupera linhas (e/ou colunas) com labels específicos.

    iloc recupera linhas (e/ou colunas) em posições inteiras específicas.

```python
df_subset = df[['year_id', 'team_id']]
df_subset
```

```python
# Recuperando elementos
print(df_subset.loc[3], df_subset.iloc[3])
```

```python
df_sorted = df_subset.sort_values(by='team_id')
df_sorted[['year_id', 'team_id']]
```

```python
df_sorted.loc[3]
```

```python
df_sorted.iloc[3]
```

# Historical NBA Elo

This directory contains the data behind the [Complete History Of The NBA](http://fivethirtyeight.com/interactives/the-complete-history-of-every-nba-team-by-elo) interactive. Data updated periodically. Game information is from [Basketball-Reference.com](http://www.basketball-reference.com/).

In the data file `nbaallelo.csv` are the following variables:

Header | Definition
---|---------
`gameorder` | Play order of game in NBA history
`game_id` | Unique ID for each game
`lg_id` | Which league the game was played in
`_iscopy` | Each row of data is tied to a single team for a single game, so _iscopy flags if this game_id has already occured for the opposing team in the same matchup
`year_id` | Season id, named based on year in which the season ended
`date_game` | Game date
`is_playoffs` | Flag for playoff games
`team_id` | Three letter code for team name, from Basketball Reference
`fran_id` | Franchise id. Multiple team_ids can fall under the same fran_id due to name changes or moves. Interactive is grouped by fran_id.
`pts` | Points scored by team
`elo_i` | Team elo entering the game
`elo_n` | Team elo following the game
`win_equiv` | Equivalent number of wins in a 82-game season for a team of elo_n quality
`opp_id` | Team id of opponent
`opp_fran` | Franchise id of opponent
`opp_pts` | Points scored by opponent
`opp_elo_i` | Opponent elo entering the game
`opp_elo_n` | Opponent elo following the game
`game_location` | Home (H), away (A), or neutral (N)
`game_result` | Win or loss for team in the team_id column
`forecast` | Elo-based chances of winning for the team in the team_id column, based on elo ratings and game location
`notes` | Additional information
