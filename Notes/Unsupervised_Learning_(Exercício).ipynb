### 2a: O dataset

- Alcohol (lcool): Quantidade de lcool presente no vinho.
- Malic acid (cido mlico): Um cido orgnico encontrado em muitas frutas e vegetais, contribui para o sabor azedo do vinho.
- Ash (Cinza): Uma medida da quantidade total de minerais presentes no vinho.
- Alcalinity of ash (Alcalinidade da cinza): Mede o quo alcalina  a cinza quando o vinho  queimado (uma maneira de avaliar os minerais).
- Magnesium (Magnsio): Quantidade deste elemento mineral no vinho.
- Total phenols (Fenis totais): Medida dos compostos fenlicos presentes, que influenciam o sabor e a cor do vinho.
- Flavanoids (Flavonoides): Um tipo especfico de fenol que tem propriedades antioxidantes.
- Non-flavanoid phenols (Fenis no flavonoides): Fenis que no so flavonoides.
- Proanthocyanins (Proantocianidinas): Um tipo de flavonoide.
- Color intensity (Intensidade de cor): Uma medida da cor do vinho.
- Hue (Matiz): Tambm relacionado  cor, descreve o tipo de vermelho do vinho (por exemplo, um matiz roxo ou laranja).
- OD280/OD315 of diluted wines (OD280/OD315 de vinhos diludos): Uma medida da densidade ptica que ajuda a determinar a concentrao de protenas no vinho.
- Proline: Um aminocido que pode afetar o sabor e a cor do vinho.
