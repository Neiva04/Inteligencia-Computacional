## Nominal
(Pode ser chamado de categórico)

Uma variável do tipo nominal pode assumir diferentes valores discretos mas não possui uma ordenação que pode ser definida de maneira natural.

- Cor do cabelo
- Cidade
- Marcas de chocolate

Resumo: Dado pode ser categorizado



## Ordinal

Os números indicam as posições relativas dos objetos, mas não a magnitude das diferenças entre eles.

*Exemplo*:
Titulação
- 1. Fundamental
- 2. Ensino Médio
- 3. Bacharelado e Licenciatura
- 4. Especialização
- 5. Mestrado
- 6. Doutorado
- 7. Pós-doutorado

Resumindo: Dado pode ser categorizada e cada elemento possui um "ranking" relativo.



## Escala intervalar (ou númerica)

Você pode categorizar, classificar e inferir intervalos iguais entre pontos de dados vizinhos, mas não há um verdadeiro ponto zero.

A diferença entre quaisquer duas temperaturas adjacentes em Celsius é a mesma: um grau. Mas zero graus é definido de forma diferente dependendo da escala - isso não significa uma ausência absoluta de temperatura.

Suponha que três alunos tenham obtido, através de um teste, suas medidas indicadoras de inteligência (QI): 130, 115 e 65. Se for verdade que o segundo aluno tem sua medida de inteligência mais próxima da do primeiro aluno do que da medida do terceiro, não é verdade que o primeiro seja duas vezes mais  inteligente do que o terceiro. [Exemplo adaptado do Prof. Moises da FGV]



Exemplos:
- Temperatura em C ou Fahrenheit
- Teste de QI (alguns falam em ordinal!)

Permite calcular a diferença entre variáveis.

Resumindo: Dado pode ser categorizada, cada elemento possui um "ranking" relativo e seus valores são espaçados de maneira igual

## Escala de Razão


Descrito de maneira clara por um professor da FGV, "Esta escala representa o nível mais rico de mensuração que se pode obter na busca do conhecimento de um objeto. Além de incorporar todas as propriedades da escala intervalar, esta escala ainda permite que se estabeleçam relações de razão e proporção entre os valores observados de suas variáveis. Isso é possível pela existência de uma origem fixa, ou zero absoluto (natural) e pela existência de uma unidade unitária de medida. A denominação dessa escala resulta do fato que, uma vez fixada sua origem, qualquer medida na escala pode ser expressa através de uma razão." [Prof. Moises da FGV]

Exemplos:
- Grande parte das medidas físicas (comprimento, peso, etc.),
- Demográficas (idade, taxas de crescimento, natalidade, mortalidade, etc.)[Prof. Moises da FGV]
- Temperatura em Kelvin

Resumindo: Dado pode ser categorizada, cada elemento possui um "ranking" relativo, seus valores são espaçados de maneira igual e tem um valor natural de zero

# Representando datas

Em qual das categorias acima a data se encaixa melhor? Depende da situação! Datas podem não se encaixar de maneira exata nas categorias acima e a forma que são representadas depende muito do contexto.

Compare os dois exemplos abaixo:

- Dias da semana  (Segunda, terça, quarta)
- Meses do ano (Janeiro, Fevereiro, Novembro)

com

- Data completa e tempo que o ticket foi criado
- Data completa e tempo que pagamento foi feito
- Anos onde ocorreu o evento X



## Diferenças

Nos exemplos citados, tanto anos quanto os timestamps estão atrelados ao conceito de intervalo. Existe claramente uma ordem entre os elementos (2018 < 2019 < 2020) e a diferença entre dois timestamps também é bem definida.

Esse tipo de dado pode ser mais diretamente utilizado para forecasting


image.png



Em relação aos meses do ano ou dias da semana, essa análise é um pouco mais delicada. Existe claramente algum nível de ordem nos elementos, mas também existe uma característica cíclica: depois de dezembro temos janeiro.

Discutimos agora duas soluções utilizadas para esses casos:

  - Tratar esses elementos como dados categóricos e aplica-se one-hot encoding (exceto em casos de árvores de decisão e derivados onde encoding, em geral, não seria necessário)
  - Tratar esses elementos como ciclos e realizar uma transformação de seus valores para que eles representem esse comportamento
