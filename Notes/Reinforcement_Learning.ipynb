# Vamos realizar um resumo da Aprendizagem por Reforço (RL)

- Pessoas e animais aprendem ao **interagir com nosso ambiente**
- Isso difere de certos outros tipos de aprendizado
   > É **ativo** em vez de passivo
   
   >As interações são frequentemente **sequenciais** - interações futuras podem depender das anteriores
- Somos orientados por **objetivos**
- Podemos **aprender** sem exemplos de comportamento ótimo
- Em vez disso, **otimizamos** algum sinal de recompensa

# O loop de aprendizado

- O agente está em um estado atual $s \in \mathcal{S}$.
- Com base no estado atual $s$, o agente seleciona uma ação $a \in \mathcal{A}$ de acordo com sua política.
- O agente executa a ação $a$ no ambiente.
- O ambiente transita para um novo estado $s' \in \mathcal{S}$ de acordo com a função de transição de estado $\mathcal{P}(s, a, s')$.
- O ambiente fornece uma recompensa imediata $r = \mathcal{R}(s, a, s')$ ao agente.
- O agente atualiza sua estimativa de valor com base na recompensa recebida e no novo estado.
- O agente repete esse processo a partir do novo estado $s'$.

# Como selecionar ações?

Depende-se de uma política $\pi$ que determina para cada estado qual ação escolher

A política $\pi$ em um problema de Aprendizado por Reforço (RL) é uma função que mapeia estados para ações. Ela determina qual ação o agente deve escolher em um determinado estado para maximizar a recompensa acumulada ao longo do tempo.

Formalmente, uma política $\pi$ é definida como uma função $\pi: \mathcal{S} \rightarrow \mathcal{A}$, onde $\mathcal{S}$ é o conjunto de estados e $\mathcal{A}$ é o conjunto de ações. Para cada estado $s \in \mathcal{S}$, a política $\pi(s)$ retorna a ação $a \in \mathcal{A}$ que o agente deve executar nesse estado.
